```python
def D_Tree(tree_size,train_X, test_X, train_y, test_y):
    D_reg = DecisionTreeRegressor(max_leaf_nodes=tree_size, random_state=1)
    D_reg.fit(train_X, train_y)
    y_pred = D_reg.predict(test_X)
    #print("RMSE:\t\t{:.4f}".format(np.sqrt(mean_squared_error(test_y, y_train_pred))))
    return y_pred
```

```python
tree_sizes = [10, 15, 25, 50, 75,85, 100, 150, 200, 300, 500,5000]
for tree_size in tree_sizes:
    y_pred = D_Tree(tree_size,train_X, test_X, train_y, test_y)
    print("Tree Size:\t\t{}\t RMSE:\t\t{:.4f}".format(tree_size,np.sqrt(mean_squared_error(test_y, y_pred))))
```

# Gradient Boosting


```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbrt = GradientBoostingRegressor(max_depth=4, n_estimators=500, random_state=1)
gbrt.fit(train_X, train_y)
```

```python
errors = [mean_squared_error(test_y, y_pred)
 for y_pred in gbrt.staged_predict(test_X)]
bst_n_estimators = np.argmin(errors)
```

```python
gbrt_best = GradientBoostingRegressor(max_depth=4,n_estimators=bst_n_estimators,random_state=1)
gbrt_best.fit(train_X, train_y)
```

```python
y_pred = gbrt_best.predict(test_X)
```

```python
print("R2 Score:\t{:.4f}".format(r2_score(test_y, y_pred)))
print("RSE:\t\t{:.4f}".format(mean_squared_error(test_y, y_pred)))
print("RMSE:\t\t{:.4f}".format(np.sqrt(mean_squared_error(test_y, y_pred))))
print("MAE:\t\t{:.4f}".format(mean_absolute_error(test_y, y_pred)))
```
